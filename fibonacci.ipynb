```python
from datetime import datetime
```

```python
# classic method (recursion)
def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n-1) + fibo(n-2)

start = datetime.now()
print(fibo(35))
duration = datetime.now() - start
print(duration.total_seconds())
```

```python
# momoized solution
memo = {1: 1, 2: 1}
def fibo2(n):
    
    if n in memo.keys():
        return memo.get(n)
    else:
       #  memo[n] = memo.get(n-1) + memo.get(n-2) <- obviously wrong!
        memo[n] = fibo2(n-1) + fibo2(n-2)
    return memo[n]

start = datetime.now()
print(fibo2(2000))
duration = datetime.now() - start
print(duration.total_seconds())
```

```python
# using just for!
def fibo3(n):
#    if n == 1 or n == 2:
#        return 1
    
    fibo = [1, 1]
    # if the list is empty python automatically passes on it !
    for i in range(2, n):
        fibo.append(fibo[i-2] + fibo[i-1])
    return fibo[n-1]

start = datetime.now()
print(fibo3(1))
duration = datetime.now() - start
print(duration.total_seconds())
```
