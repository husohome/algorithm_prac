```python
# 9 = wall,
# 0 = init,
# 1 = path
# 2 = walked
# 3 = goal

maze = [
    [9, 9, 9, 9, 9, 9, 9],
    [9, 1, 1, 1, 9, 1, 9],
    [9, 1, 9, 1, 9, 1, 9],
    [9, 1, 9, 1, 1, 1, 9],
    [9, 1, 9, 1, 1, 1, 9],
    [9, 1, 9, 1, 9, 9, 9],
    [9, 9, 1, 1, 9, 1, 9],
    [9, 1, 1, 1, 9, 1, 9],
    [9, 1, 9, 1, 1, 1, 9],
    [9, 9, 1, 1, 9, 1, 9],
    [9, 1, 1, 9, 9, 1, 9],
    [9, 1, 9, 1, 9, 1, 9],
    [9, 9, 9, 9, 9, 3, 9]
]

m, n = len(maze), len(maze[0])
```

```python
def visit(x, y):
    if maze[x][y] == 3:
        return maze
   
    if maze[x][y] == 1:
        maze[x][y] = 0
    
    if maze[x+1][y] == 1:
        visit(x+1, y) 
    
    if maze[x-1][y] == 1:
        visit(x-1, y)
    
    if maze[x][y+1] == 1:
        visit(x, y+1) 
    
    if maze[x][y-1] == 1:
        visit(x, y-1) 
```

```python
visit(1,1)
for row in maze:
    print(row)
```
