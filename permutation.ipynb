```python
a = list("abcd")

def perm(k, a):
    if k == 1:
        return a
    else:
        perm(k-1, a)
        
        for i in range(k):
            if k % 2 == 0:
                a[i], a[k-1] = a[k-1], a[i]
            else:
                a[0], a[k-1] = a[k-1], a[0]
            yield perm(k-1, a)

for i in perm(3, a):
    print(i)
```
